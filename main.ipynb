```python
from philly_cheesesnake.models import Resource

resources = [
    Resource.from_dict(r) for rs in cs.datasets_df.resources if rs for r in rs if r
]
print(len(resources))
print(resources[:3])
```

```python
import re
from collections import defaultdict

grouped = defaultdict(set)
for r in resources:
    fmt = r.format.lower()
    name = re.sub(rf"\s*\({fmt}\)$", "", r.name, flags=re.IGNORECASE).strip()
    grouped[name].add(fmt)

# All formats seen
all_formats = sorted({fmt for fmts in grouped.values() for fmt in fmts})

# Build table with ✅
rows = []
for name, fmts in grouped.items():
    row = {"name": name}
    for fmt in all_formats:
        row[fmt] = "✅" if fmt in fmts else ""
    rows.append(row)
```

```python
import pandas as pd

resources_df = pd.DataFrame([r for r in rows])
pd.set_option("display.max_rows", None, "display.max_columns", None)
resources_df
```
